```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('datasets_4458_8204_winequality-red.csv')
df.head()
```

```python
sns.set(style="white")

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(df.corr(), cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
qual_corr = df.corr().sort_values('quality', ascending=False)['quality']
print(qual_corr)
```

It seems the features that have the most correlation to quality are alcohol, volatile acidity, sulphates, citric acid

```python
df_most_corr = df[['alcohol', 'volatile acidity', 'sulphates', 'citric acid']]
for i in df_most_corr:
    plt.scatter(df[i], df.quality)
    plt.title(f"Scatter plot of quality vs {i}")
    plt.xlabel(i)
    plt.ylabel('Quality')
    plt.show()
```
